```python
#Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load CSV data into dataframe

df = pd.read_csv("top2018.csv")
df.head()
```

```python
#Checking Data Structure

df.info()
df.describe(include='all')
```

```python
#Check for missing values

df.isna().sum()
```

```python
#Check for duplicated values

df.duplicated().sum()
```

```python
#Changing the duration column from milliseconds to minutes

df['duration'] = (df['duration_ms'] / 60000).round(2)
df.drop(columns=['duration_ms'], inplace=True)
```

```python
df.head()
```

```python
#Distribution of Danceability

sns.histplot(df['danceability'], kde=True, color='purple')
plt.title("Distribution of Danceability")
plt.show()
```

```python
#Top Artist analysis

top_artists = df['artists'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_artists.values, y=top_artists.index)
plt.title("Top Artists with Most Tracks in 2018")
plt.xlabel("Number of Tracks")
plt.ylabel("Artist")
plt.show()
```

```python
#Energy vs Danceability

sns.scatterplot(data=df, x='energy', y='danceability', color="red")
plt.title("Energy vs Danceability")
plt.show()
```

```python
#Correlation Matrix
#Code has been written to disregard columns with string values

plt.figure(figsize=(12,8))

numeric_df = df.select_dtypes(include=['float64', 'int64'])

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Audio Features")
plt.show()
```

```python
#This graph is to check if hit songs are getting faster?

plt.figure(figsize=(10,5))
sns.histplot(df['tempo'], kde=True)
plt.title("Distribution of Song Tempo (BPM)")
plt.show()
```

```python
#Distribution of song duration

plt.figure(figsize=(10,5))
sns.histplot(df['duration'], kde=True)
plt.title("Distribution of Song Duration (ms)")
plt.show()
```

```python
#Song count by Key

df['key'].value_counts().head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top Genres in the Dataset")
plt.xlabel("Key")
plt.ylabel("Count")
plt.show()
```
